```python
import eyeloc as el
```

```python
finder_p = el.PolarFinder(el.Find.Pupil, el.CenterEstimation.LargestComponentCentroid, filter_model=el.CircleRANSAC(2, 70, 100))
finder_i = el.PolarFinder(el.Find.Iris, el.CenterEstimation.LargestComponentCentroid, filter_model=el.CircleRANSAC(2, 70, 100))
```

```python
left, right = el.eyeimread("test_images/eyes1.png")
```

```python
image = left
center_p, radius_p = finder_p.find(image)
center_i, radius_i = finder_i.find(image)
el.show_circles(image, (center_p, center_i), (radius_p, radius_i), [(255, 0, 0), (0, 255, 0)])
```
